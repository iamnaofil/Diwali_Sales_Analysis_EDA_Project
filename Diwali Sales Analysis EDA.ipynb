**All Keys Finding:**<br>


**Gender Disparity:** Females are the dominant purchasers, outnumbering male buyers significantly.<br>

**Age Group:** The 26-35 age group of females constitutes the largest customer segment, with 3269 purchases, while males in the same age group made 1272 purchases.<br>

**Regional Variation:** Uttar Pradesh, Maharashtra, and Karnataka lead in the number of orders, indicating a higher demand for products in these regions.<br>

**Marital Status:** Married individuals, particularly females, are more inclined to spend larger amounts.<br>

**Sector Influence:** The IT sector emerges as a significant driver of sales, with higher purchasing power compared to the Healthcare sector.<br>

**Popular Products:** Clothing products are the most frequently ordered, with foods being the top-selling category.<br>

**Spending Patterns:** Spending on Food, Clothing & Apparel, and Household Items is nearly equal between genders. However, males tend to spend slightly more on Hands and Power Tools.<br>

**Top Products:** The top 5 selling products are identified as p00265242, p00110942, p00237542, p00184942, and p00114942.<br>

# **Conclusion**

The analysis reveals that married women in the 26-35 age group, primarily located in Uttar Pradesh, Maharashtra, and Karnataka, working in the IT, Healthcare, and Aviation sectors, exhibit a strong preference for products in the Food, Clothing, and Electronics categories. These insights provide valuable guidance for businesses aiming to target specific customer segments and product categories during Diwali sales
